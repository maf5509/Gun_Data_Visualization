```python
y2012 = gun_d['year'] == 2012# & ~np.isnan(gun_d['age'])
gun_d2012 = gun_d[y2012]
gun_d2012fin = ~np.isnan(gun_d2012['age'])
gun_d2012final = gun_d2012[gun_d2012fin]

y2013 = gun_d['year'] == 2013# & ~np.isnan(gun_d['age'])
gun_d2013 = gun_d[y2013]
gun_d2013fin = ~np.isnan(gun_d2013['age'])
gun_d2013final = gun_d2013[gun_d2013fin]

y2014 = gun_d['year'] == 2014# & ~np.isnan(gun_d['age'])
gun_d2014 = gun_d[y2014]
gun_d2014fin = ~np.isnan(gun_d2014['age'])
gun_d2014final = gun_d2014[gun_d2014fin]

gun_d2012final.head()
```

```python
plt.figure(figsize=(18, 3))

plt.subplot(1, 3, 1)

sns.distplot(gun_d2012final['age'], bins=10)#, kde=False, rug=True)
plt.xlabel('Age of Victim')
plt.ylabel('Frequency')
plt.title('2012 US Gun Deaths - Age Distribution')

plt.subplot(1,3,2)

sns.distplot(gun_d2013final['age'], bins=10)#, kde=False, rug=True)
plt.xlabel('Age of Victim')
plt.ylabel('Frequency')
plt.title('2013 US Gun Deaths - Age Distribution')

plt.subplot(1,3,3)

sns.distplot(gun_d2014final['age'], bins=10)#, kde=False, rug=True)
plt.xlabel('Age of Victim')
plt.ylabel('Frequency')
plt.title('2014 US Gun Deaths - Age Distribution')

plt.show()
```

```python
yacc = gun_d['intent'] == 'Accidental'
gun_dacc = gun_d[yacc]
gun_daccfin = ~np.isnan(gun_dacc['age'])
gun_daccfinal = gun_dacc[gun_daccfin]

yhom = gun_d['intent'] == 'Homicide'
gun_dhom = gun_d[yhom]
gun_dhomfin = ~np.isnan(gun_dhom['age'])
gun_dhomfinal = gun_dhom[gun_dhomfin]

ysui = gun_d['intent'] == 'Suicide'
gun_dsui = gun_d[ysui]
gun_dsuifin = ~np.isnan(gun_dsui['age'])
gun_dsuifinal = gun_dsui[gun_dsuifin]

yund = gun_d['intent'] == 'Undetermined'
gun_dund = gun_d[yund]
gun_dundfin = ~np.isnan(gun_dund['age'])
gun_dundfinal = gun_dund[gun_dundfin]

gun_dhomfinal.head()
```

```python
plt.figure(figsize=(18, 10))

#distplot plots kernel density estimate by default. for reg histogram set to False
plt.subplot(2,2,1)
sns.distplot(gun_daccfinal['age'], bins=10)#, kde=True, rug=True)
plt.xlabel('Age of Victim')
plt.ylabel('Frequency')
plt.title('US Accidental Gun Deaths - Age Distribution')

plt.subplot(2,2,2)
sns.distplot(gun_dhomfinal['age'], bins=10)#, kde=True, rug=True)
plt.xlabel('Age of Victim')
plt.ylabel('Frequency')
plt.title('US Homicide Gun Deaths - Age Distribution')

plt.subplot(2,2,3)
sns.distplot(gun_dsuifinal['age'], bins=10)#, kde=True, rug=True)
plt.xlabel('Age of Victim')
plt.ylabel('Frequency')
plt.title('US Suicide Gun Deaths - Age Distribution')

plt.subplot(2,2,4)
sns.distplot(gun_dundfinal['age'], bins=10)#, kde=True, rug=True)
plt.xlabel('Age of Victim')
plt.ylabel('Frequency')
plt.title('US Gun Deaths (Manner Undetermined) - Age Distribution')

plt.show()
```
